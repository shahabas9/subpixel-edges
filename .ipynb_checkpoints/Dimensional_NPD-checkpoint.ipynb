# Minimum tolerance code for diameter

```python
image_directory = "/home/shahabas/SAC/Dimension_New/22_june_23/Same_job_moving"
filenames = os.listdir(image_directory)

for file in filenames[:]:
    input_path = os.path.join(image_directory, file)
    image = cv2.imread(input_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    croped_image = gray[1200:1400]

    # Apply Canny edge detection to find the edges
    edges = cv2.Canny(croped_image, 80, 200)
    
    # Apply the Hough line transform
    lines = cv2.HoughLines(edges, 1, np.pi/180, threshold=90)

    # Extract the data points from the detected lines
    data_points = []
    for line in lines:
        rho, theta = line[0]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))
        data_points.append([(x1 + x2) // 2, (y1 + y2) // 2])

    # Convert the data points to NumPy array
    data_points = np.array(data_points)

    # Separate the x and y coordinates
    x = data_points[:, 0]
    y = data_points[:, 1]
    print(y)
    # Perform linear regression
    slope, intercept = np.polyfit(x, y, 1)

    # Generate points for the best fit line
    best_fit_line_x = np.linspace(min(x), max(x), 100)
    best_fit_line_y = slope * best_fit_line_x + intercept
    
    # Generate points for the best fit line
    plt.imshow(cv2.cvtColor(croped_image, cv2.COLOR_BGR2RGB))
    plt.plot(best_fit_line_x, best_fit_line_y, color='yellow', linewidth=2)
    plt.scatter(x, y, color='red', marker='x')
    plt.axis('off')
    plt.show()
    print((max(x)-min(x))/41.45)
```

# Finalized code for dia

```python
import time
image_directory = "/home/shahabas/SAC/Dimension_New/22_june_23/Same_job_moving"
filenames = os.listdir(image_directory)
data=[]
for file in filenames:
    start_time=time.time()
    input_path = os.path.join(image_directory, file)
    image = cv2.imread(input_path)
    cropped_image = image[1200:1400]
    gray = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)
    _, thresholded = cv2.threshold(gray, 20, 255, cv2.THRESH_BINARY_INV)

    height, width = thresholded.shape[:2]

    kernel_size = 5
    blur_gray = cv2.GaussianBlur(thresholded, (kernel_size, kernel_size), 0)

    low_threshold = 30
    high_threshold = 90
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)

    lines = cv2.HoughLinesP(edges, rho=1, theta=np.pi/180, threshold=10, minLineLength=60, maxLineGap=35)
#     print("Number of lines:", len(lines))
    
    x_left=[]
    x_right=[]
    for line in lines:
        x1, y1, x2, y2 = line[0]
#         print(x1,y1,x2,y2)
        if x1<1000 or x2<1000:
            x_left.append(x1)
            x_left.append(x2)
        else:
            x_right.append(x1)
            x_right.append(x2)
    x_org_left=int(np.mean(x_left))
    x_org_right=int(np.mean(x_right))
    diameter=(x_org_right-x_org_left)/41.45
    print(str(diameter)+" mm")
    three_channel_image = np.zeros((height, width, 3), dtype=np.uint8)
    thrbee_channel_image[:, :, 0] = thresholded
    three_channel_image[:, :, 1] = thresholded
    three_channel_image[:, :, 2] = thresholded

    inverted_image = cv2.bitwise_not(thresholded)
    colour_inverted=cv2.cvtColor(inverted_image, cv2.COLOR_BGR2RGB)
    cv2.line(colour_inverted,(x_org_left,0),(x_org_left,height),(255,0,0),3)
    cv2.line(colour_inverted,(x_org_right,0),(x_org_right,height),(255,0,0),4)
    cv2.line(colour_inverted,(x_org_left,height//2),(x_org_right,height//2),(0,255,0),3)
    end_time=time.time()
    total_time=end_time-start_time
    print("the total time is ",total_time*1000,"millisecond")
    print(file)
    show_image(colour_inverted)
    data.append([file,str(diameter)+" mm",total_time*1000])
```
